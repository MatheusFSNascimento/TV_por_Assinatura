```python
#Importando as bibliotecas necessárias
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.patheffects as pe
import seaborn as sns
```

```python
#Carregando a base de dados
dataset = pd.read_csv('Acessos_TV_Assinatura.csv', sep=';')
```

```python
municipios = dataset.Município.unique()
municipios
```

```python
acessosmunicipio = {}

for municipio in municipios:
    acessosmunicipio[municipio] = dataset[dataset['Município']==municipio].Acessos.sum()
```

```python
acessosmunicipio
```

```python
top10_municipios = sorted(acessosmunicipio.keys(), key= acessosmunicipio.get, reverse=True)[:10]
top10_municipios
```

```python
top10_valores = sorted(acessosmunicipio.values(), reverse=True)[:10]
top10_valores
```

```python
dataset_top10municipios = pd.DataFrame(top10_municipios)
dataset_top10municipios
```

```python
dataset_top10valores = pd.DataFrame(top10_valores)
dataset_top10valores
```

```python
dataset_final = pd.concat([dataset_top10municipios, dataset_top10valores], axis=1)
dataset_final
```
